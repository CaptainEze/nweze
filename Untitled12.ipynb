```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Load sample dataset
pressure = dseats_2024_training["FP"]
# Split Dataset 1
X1_train, X1_val, y1_train, y1_val = train_test_split(dseats_2024, pressure, test_size=0.2, random_state=42)

# Decision Tree Model
tree_model = DecisionTreeRegressor(random_state=42)
tree_model.fit(X1_train, y1_train)
tree_pred = tree_model.predict(X1_val)

# Evaluate Decision Tree Model
tree_mse = mean_squared_error(y1_val, tree_pred)
tree_r2 = r2_score(y1_val, tree_pred)

print(f"Decision Tree MSE: {tree_mse:.2f}")
print(f"Decision Tree R^2 Score: {tree_r2:.2f}")

# Random Forest Model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X1_train, y1_train)
rf_pred = rf_model.predict(X1_val)

# Evaluate Random Forest Model
rf_mse = mean_squared_error(y1_val, rf_pred)
rf_r2 = r2_score(y1_val, rf_pred)

print(f"Random Forest MSE: {rf_mse:.2f}")
print(f"Random Forest R^2 Score: {rf_r2:.2f}")

# ANN Model
ann_model = Sequential([
    Dense(64, input_shape=(X1_train.shape[1],), activation='relu'),
    Dense(32, activation='relu'),
    Dense(1)  # Output layer for regression
])
ann_model.compile(optimizer=Adam(learning_rate=0.01), loss='mse')

# Train ANN model
ann_model.fit(X1_train, y1_train, epochs=100, batch_size=10, verbose=0)

# Predict with ANN model
ann_pred = ann_model.predict(X1_val).flatten()

# Evaluate ANN Model
ann_mse = mean_squared_error(y1_val, ann_pred)
ann_r2 = r2_score(y1_val, ann_pred)

print(f"ANN MSE: {ann_mse:.2f}")
print(f"ANN R^2 Score: {ann_r2:.2f}")

# Summary of Results
results = pd.DataFrame({
    'Model': ['Decision Tree', 'ANN', 'Random Forest'],
    'MSE': [tree_mse, rf_mse, ann_mse],
    'R^2 Score': [tree_r2, rf_r2, ann_r2]
})
print("\nComparison of Models:")
print(results)
```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow import keras
import matplotlib.pyplot as plt

import seaborn as sns

# Load the dataset (assuming you have it as a CSV)
# Replace 'dataset.csv' with the actual dataset path
# Assuming you have three datasets: X1, X2, X3 (features) and y1, y2, y3 (target variables)
#Feature
pressure = dseats_2024_training["FP"]
# Split Dataset 1
X1_train, X1_val, y1_train, y1_val = train_test_split(dseats_2024, pressure, test_size=0.2, random_state=42)

print("Dataset 1:")
print("Train set X1:", X1_train.shape, "Train set y1:", y1_train.shape)
print("Test set X1:", X1_val.shape, "Test set y1:", y1_val.shape)


# Scale the features (normalization/standardization)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X1_train)
X_test_scaled = scaler.transform(X1_val)


# Build the ANN model using Functional API
# Input layer
input_layer = keras.layers.Input(shape=X_train_scaled.shape[1:])

# First hidden layer with 30 neurons and ReLU activation
hidden1 = keras.layers.Dense(30, activation='relu')(input_layer)

# Second hidden layer with 30 neurons and ReLU activation
hidden2 = keras.layers.Dense(30, activation='relu')(hidden1)

# Concatenation of input and second hidden layer
concat = keras.layers.Concatenate()([input_layer, hidden2])

# Output layer (single neuron for regression task)
output_layer = keras.layers.Dense(1)(concat)

# Create the model
model = keras.models.Model(inputs=[input_layer], outputs=[output_layer])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_squared_error'])

# Train the model
history = model.fit(X_train_scaled, y1_train, epochs= 500, batch_size=32, validation_split=0.2)

# Plot training history (loss)
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Model Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.show()

# Make predictions on the test set
y_pred = model.predict(X_test_scaled)

# Model performance evaluation
mae = mean_absolute_error(y1_val, y_pred)
r2 = r2_score(y1_val, y_pred)

print(f'Mean Absolute Error: {mae}')
print(f'R-squared: {r2}')

```

```python
import scipy.stats as stats

# Contingency tables for breed and market
breed_data = [[20, 30], [17, 42]]  # [Positive, Negative] for Local and Exotic breeds
market_data = [[23, 27], [14, 36]]  # [Positive, Negative] for Rumuokoro and Choba markets

# Chi-square test for breed
chi2_breed, p_breed, _, _ = stats.chi2_contingency(breed_data)
print(f"Chi-square test for breed: Chi2 = {chi2_breed:.2f}, p-value = {p_breed:.4f}")
if p_breed < 0.05:
    print("There is a significant difference in prevalence based on breed.")
else:
    print("No significant difference in prevalence based on breed.")

# Chi-square test for market
chi2_market, p_market, _, _ = stats.chi2_contingency(market_data)
print(f"Chi-square test for market: Chi2 = {chi2_market:.2f}, p-value = {p_market:.4f}")
if p_market < 0.05:
    print("There is a significant difference in prevalence based on market.")
else:
    print("No significant difference in prevalence based on market.")
```

```python
import scipy.stats as stats

# Observed values: 37 positive and 63 negative out of 100 samples
observed_overall = [37, 63]

# Expected values assuming a 50-50 distribution
expected_overall = [50, 50]

# Perform Chi-square goodness-of-fit test
chi2_overall, p_overall = stats.chisquare(f_obs=observed_overall, f_exp=expected_overall)
print(f"Chi-square test for overall prevalence: Chi2 = {chi2_overall:.2f}, p-value = {p_overall:.4f}")

# Interpretation of p-value
if p_overall < 0.05:
    print("There is a significant difference in overall prevalence compared to the expected distribution.")
else:
    print("No significant difference in overall prevalence compared to the expected distribution.")
```
